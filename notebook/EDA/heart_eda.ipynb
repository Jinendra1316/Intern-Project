```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

pd.set_option("display.max_columns",None)
```

```python
df = pd.read_csv("../../data/processed/heart_processed.csv")

print(df.shape)

display(df.head())
```

```python
df.info()
```

```python
display(df.describe())
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="cardio",data=df)

plt.title("Cardiovascular Disease Distribution")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="gender",data=df)

plt.title("Gender Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["age"],bins=30,kde=True)

plt.title("Age Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["height"],bins=30,kde=True)

plt.title("Height Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["weight"],bins=30,kde=True)

plt.title("Weight Distribution")

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)

sns.histplot(df["ap_hi"],bins=30,kde=True)

plt.title("Systolic Blood Pressure")

plt.subplot(1,2,2)

sns.histplot(df["ap_lo"],bins=30,kde=True)

plt.title("Diastolic Blood Pressure")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="cholesterol",data=df)

plt.title("Cholesterol Levels")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="gluc",data=df)

plt.title("Glucose Levels")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="smoke",data=df)

plt.title("Smoking")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="active",data=df)

plt.title("Physical Activity")

plt.show()
```

```python
numerical_columns=[
    "age",
    "height",
    "weight",
    "ap_hi",
    "ap_lo"
]

for column in numerical_columns:

    plt.figure(figsize=(7,4))

    sns.boxplot(x=df[column])

    plt.title(column)

    plt.show()
```

```python
plt.figure(figsize=(12,10))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="cardio",
    y="weight",
    data=df
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="cardio",
    y="age",
    data=df
)

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)

sns.boxplot(
    x="cardio",
    y="ap_hi",
    data=df
)

plt.subplot(1,2,2)

sns.boxplot(
    x="cardio",
    y="ap_lo",
    data=df
)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x="gluc",
    hue="cardio",
    data=df
)

plt.show()
```

```python
sns.pairplot(
    df[
        [
            "age",
            "weight",
            "ap_hi",
            "ap_lo",
            "cardio"
        ]
    ],
    hue="cardio"
)

plt.show()
```
