```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../../data/processed/kidney_processed.csv")

print(df.shape)

display(df.head())
```

```python
df.info()
```

```python
display(df.describe())
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="Diagnosis", data=df)

plt.title("Kidney Disease Distribution")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="Gender", data=df)

plt.title("Gender Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"], bins=30, kde=True)

plt.title("Age Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["BMI"], bins=30, kde=True)

plt.title("BMI Distribution")

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)

sns.histplot(df["SystolicBP"], bins=30, kde=True)

plt.title("Systolic Blood Pressure")

plt.subplot(1,2,2)

sns.histplot(df["DiastolicBP"], bins=30, kde=True)

plt.title("Diastolic Blood Pressure")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)

sns.histplot(df["FastingBloodSugar"], bins=30, kde=True)

plt.title("Fasting Blood Sugar")

plt.subplot(1,2,2)

sns.histplot(df["HbA1c"], bins=30, kde=True)

plt.title("HbA1c")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)

sns.histplot(df["SerumCreatinine"], kde=True)

plt.title("Serum Creatinine")

plt.subplot(1,3,2)

sns.histplot(df["BUNLevels"], kde=True)

plt.title("BUN Levels")

plt.subplot(1,3,3)

sns.histplot(df["GFR"], kde=True)

plt.title("GFR")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)

sns.histplot(df["ProteinInUrine"], kde=True)

plt.title("Protein in Urine")

plt.subplot(1,2,2)

sns.histplot(df["ACR"], kde=True)

plt.title("Albumin Creatinine Ratio")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.histplot(df["HemoglobinLevels"], kde=True)

plt.title("Hemoglobin Levels")

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)

sns.histplot(df["CholesterolTotal"], kde=True)

plt.title("Total Cholesterol")

plt.subplot(1,3,2)

sns.histplot(df["CholesterolLDL"], kde=True)

plt.title("LDL")

plt.subplot(1,3,3)

sns.histplot(df["CholesterolHDL"], kde=True)

plt.title("HDL")

plt.tight_layout()

plt.show()
```

```python
columns = [
    "Age",
    "BMI",
    "SystolicBP",
    "DiastolicBP",
    "SerumCreatinine",
    "BUNLevels",
    "GFR"
]

for col in columns:

    plt.figure(figsize=(7,4))

    sns.boxplot(x=df[col])

    plt.title(col)

    plt.show()
```

```python
plt.figure(figsize=(16,14))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Diagnosis",
    y="Age",
    data=df
)

plt.show()
```
