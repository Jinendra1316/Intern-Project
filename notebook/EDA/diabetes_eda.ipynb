```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../../data/processed/diabetes_processed.csv")

print(df.shape)

display(df.head())
```

```python
df.info()
```

```python
display(df.describe())
```

```python
plt.figure(figsize=(7,5))

sns.countplot(x="diabetes", data=df)

plt.title("Distribution of Diabetes")
plt.xlabel("Diabetes")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(x="gender", data=df)

plt.title("Gender Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x="smoking_history", data=df)

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["age"], bins=30, kde=True)

plt.title("Age Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["bmi"], bins=30, kde=True)

plt.title("BMI Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["HbA1c_level"], bins=30, kde=True)

plt.title("HbA1c Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["blood_glucose_level"], bins=30, kde=True)

plt.title("Blood Glucose Level")

plt.show()
```

```python
numerical_columns = [
    "age",
    "bmi",
    "HbA1c_level",
    "blood_glucose_level"
]

for column in numerical_columns:

    plt.figure(figsize=(7,4))

    sns.boxplot(x=df[column])

    plt.title(column)

    plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="diabetes",
    y="age",
    data=df
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="diabetes",
    y="bmi",
    data=df
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="diabetes",
    y="blood_glucose_level",
    data=df
)

plt.show()
```

```python
sns.pairplot(
    df[
        [
            "age",
            "bmi",
            "HbA1c_level",
            "blood_glucose_level",
            "diabetes"
        ]
    ],
    hue="diabetes"
)

plt.show()
```
