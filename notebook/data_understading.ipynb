```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
DIABETES_PATH = "../data/raw/diabetes_prediction_dataset.csv"
HEART_PATH = "../data/raw/cardio_train.csv"
KIDNEY_PATH = "../data/raw/Chronic_Kidney_Disease_data.csv"
```

```python
diabetes_df = pd.read_csv(DIABETES_PATH)

heart_df = pd.read_csv(
    HEART_PATH,
    sep=";"
)

kidney_df = pd.read_csv(KIDNEY_PATH)
```

```python
print("Diabetes Dataset Shape :", diabetes_df.shape)

print("Heart Dataset Shape :", heart_df.shape)

print("Kidney Dataset Shape :", kidney_df.shape)
```

```python
print("Diabetes Dataset")
display(diabetes_df.head())

print("Heart Dataset")
display(heart_df.head())

print("Kidney Dataset")
display(kidney_df.head())
```

```python
print("Diabetes Columns")
print(diabetes_df.columns.tolist())

print("\nHeart Columns")
print(heart_df.columns.tolist())

print("\nKidney Columns")
print(kidney_df.columns.tolist())
```

```python
print("========== Diabetes ==========")
diabetes_df.info()

print("\n========== Heart ==========")
heart_df.info()

print("\n========== Kidney ==========")
kidney_df.info()
```

```python
print("Diabetes")
display(diabetes_df.dtypes)

print("Heart")
display(heart_df.dtypes)

print("Kidney")
display(kidney_df.dtypes)
```

```python
print("Diabetes")
display(diabetes_df.describe(include="all"))

print("Heart")
display(heart_df.describe(include="all"))

print("Kidney")
display(kidney_df.describe(include="all"))
```

```python
print("Diabetes Missing Values")
display(diabetes_df.isnull().sum())

print("\nHeart Missing Values")
display(heart_df.isnull().sum())

print("\nKidney Missing Values")
display(kidney_df.isnull().sum())
```

```python
print("Diabetes Duplicates :", diabetes_df.duplicated().sum())

print("Heart Duplicates :", heart_df.duplicated().sum())

print("Kidney Duplicates :", kidney_df.duplicated().sum())
```

```python
for name, df in {
    "Diabetes": diabetes_df,
    "Heart": heart_df,
    "Kidney": kidney_df
}.items():
    
    print("="*60)
    print(name)
    print("="*60)

    display(df.nunique().sort_values())
```

```python
print("Diabetes Target")
display(diabetes_df.iloc[:, -1].value_counts())

print("\nHeart Target")
display(heart_df.iloc[:, -1].value_counts())

print("\nKidney Target")
display(kidney_df.iloc[:, -1].value_counts())
```

```python
summary = pd.DataFrame({
    "Dataset": ["Diabetes", "Heart", "Kidney"],
    "Rows": [
        diabetes_df.shape[0],
        heart_df.shape[0],
        kidney_df.shape[0]
    ],
    "Columns": [
        diabetes_df.shape[1],
        heart_df.shape[1],
        kidney_df.shape[1]
    ],
    "Missing Values": [
        diabetes_df.isnull().sum().sum(),
        heart_df.isnull().sum().sum(),
        kidney_df.isnull().sum().sum()
    ],
    "Duplicate Rows": [
        diabetes_df.duplicated().sum(),
        heart_df.duplicated().sum(),
        kidney_df.duplicated().sum()
    ]
})

display(summary)
```
